# Trait Adjective Data

To make sure each adjective is used for only one trait, I use this notebook to find duplicates.

```python
import pandas as pd
```

```python
!git clone https://github.com/DV-11/SpanishDialectDiscrimination.git

```

```python
df = pd.read_csv("/content/SpanishDialectDiscrimination/Data/adjective_dataset_v1.csv")
```

```python
df.head()
```

# Check for duplitactes between columns

```python
all_adjs = set()
for i in df.columns:
  for j in df[i].unique():
    if j is not None:
      all_adjs.add(j)
```

```python
repeated_adjs = {}

for i in all_adjs:
  cols = df.columns[df.isin([i]).any()].tolist()
  if len(cols) > 1 and type(i) is not float:
    repeated_adjs[i] = cols
```

```python
repeated_adjs
```

The cell above shows which adjectives appear in multiple columns. The column with the adjective in the highest row keeps it.
