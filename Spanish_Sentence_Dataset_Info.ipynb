```python
import pandas as pd
```

```python
di = pd.read_csv('dialogue_info.csv')
di.head()
```

```python
ss = pd.read_csv('spanish_sentences.csv')
ss.head()
```

```python
id_count = ss['Text ID'].value_counts().to_dict()

mx_ids = set(di[di['Country']=='Mexico']['Text_ID'])

es_ids = mex_ids = set(di[di['Country']=='Spain']['Text_ID'])
```

```python
mx_id_counts = {k: id_count[k] for k in mx_ids}
es_id_counts = {k: id_count[k] for k in es_ids}
```

```python
print("Number of sentences originally from Peninsular Spanish texts:", sum(es_id_counts.values()))
print("Number of sentences originally from Mexican Spanish texts:", sum(mx_id_counts.values()))
```
